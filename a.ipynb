```python
import geopandas as gpd

import matplotlib.pyplot as plt

# Load the GeoJSON file
geo_data = gpd.read_file(r"Former Municipality Boundaries.geojson")

# Plot the borders
geo_data.boundary.plot(color='black', linewidth=1, figsize=(10, 8), aspect=1)
plt.title("Old City Borders")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
