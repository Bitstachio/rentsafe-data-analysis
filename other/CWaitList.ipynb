Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"data/Centralized Waitlist for Social Housing.csv")
```

```python
data
```

Wait List Over Time

```python
# Convert 'Total active waiting list' to numeric by removing commas
new_dataframe = data[['Quarter/Year', 'Total active waiting list']].copy()
new_dataframe['Total active waiting list'] = new_dataframe['Total active waiting list'].str.replace(',', '').astype(int)

new_dataframe.head()
```

```python
# Create a line plot to compare Quarter/Year with Total active waiting list
plt.figure(figsize=(10, 6))
sns.lineplot(data=new_dataframe, x='Quarter/Year', y='Total active waiting list', marker='o')

# Customize the plot
plt.title('Total Active Waiting List Over Time', fontsize=16)
plt.xlabel('Quarter/Year', fontsize=12)
plt.ylabel('Total Active Waiting List', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

Special Priority

```python
new_dataframe = data[['Quarter/Year', 'Special Priority']]
sns.lineplot(data=new_dataframe, x='Quarter/Year', y='Special Priority', marker='o')

# Customize the plot
plt.title('Special Priority Over Time', fontsize=16)
plt.xlabel('Quarter/Year', fontsize=12)
plt.ylabel('Total Active Waiting List', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```
